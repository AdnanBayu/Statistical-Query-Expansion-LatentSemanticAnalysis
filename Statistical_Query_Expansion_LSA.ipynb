```python
from bs4 import BeautifulSoup
import requests
import re

def open_article(url):
    page = requests.get(url)
    soup = BeautifulSoup(page.content,"html.parser")
    return soup

def get_element(post):
    data = {}
    data["article_url"] = post.find('a')['href']

    article_page = open_article(data["article_url"])
    article_title = article_page.find("h1").text
    data['title'] = article_title
    
    article_content = article_page.find("div", {'itemprop':'description articleBody'})
    # content_divs = article_content.find_all("div")[1]
    
    # Assuming the relevant content is in the first div, adjust as needed
    if article_content:
        content = article_content.text.replace('\n',"")
        # print(content)
        
        match = re.search(r"[PERTAANYAAN|PERTANYAAN](.*)JAWABAN(.*)", content)
        
        if match:
            data['judul'] = data['title']
            data['pertanyaan'] = match[1].replace("ERTANYAAN","")
            data['jawaban'] = match[2]
            data['link'] = data["article_url"]
        else:
            data['pertanyaan'] = "Pertanyaan tidak ditemukan."
            data['jawaban'] = "Jawaban tidak ditemukan."
            data['link'] = data["article_url"]
            data['judul'] = data['title']
    else:
        data['pertanyaan'] = "Pertanyaan tidak ditemukan."
        data['jawaban'] = "Jawaban tidak ditemukan."
        data['link'] = data["article_url"]
        data['judul'] = data['title']
        # data['content'] = None

    return data

datas = []

for url in list_url_base:


    page = open_article(url)

    posts = page.find_all("div",class_="post-body entry-content")

    for post in posts:
        data = get_element(post)
        
        print("Judul : ", data["title"])
        print("Pertanyaan : ", data["pertanyaan"])
        print("Jawaban : ", data["jawaban"])
        print("Link : ", data["link"])
        # print("Content :", data['content'])
        datas.append(data)
        print("------------------------")
```

```python
def remove_chars(text):
    remove_stuffs = ['\n','+','&','%','$','@','#','!','[',']','^','-','=',":"]
    return ''.join(char for char in text if char not in remove_stuffs)
```

```python
def extract_text(text):
    text = text.lower()
    
    pattern = r'^[0-9. "]{4}(.*)$'
    match = re.match(pattern, text)
    if match:
        return match.group(1)
    else:
        return text
```

```python
df['judul'] = df['judul'].apply(extract_text)
df.head()
```

```python
def remove_two_first(text):
    cleaned = text[2:]
    return ''.join(cleaned)
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
stop_factory = StopWordRemoverFactory()

stopwords = stop_factory.get_stop_words()
```

```python
def stopword_filter(sentences):
    filtered_sentences = []
    for sentence in sentences:
        filtered = [word for word in sentence.lower().split() if word not in stopwords]
        filtered_sentences.append(' '.join(filtered))
        
    return filtered_sentences
```

```python
import re

def clean_text(text):
    # Remove HTML tags, special characters, and control characters
    text = re.sub(r"<[^>]+>", "", text)  # Remove HTML tags
    text = re.sub(r"\\u[0-9a-fA-F]{4}", "", text)  # Remove Unicode escape sequences
    text = re.sub(r"[^\w\s\.\?!]", "", text)  # Remove non-alphanumeric, whitespace, or punctuation

    # Clean up individual sentences (modified)
    clean_sentences = []
    for sentence in re.split(r"[.?!]+", text):  # Split on punctuation (one or more times)
        sentence = sentence.strip()  # Remove leading/trailing whitespace
        sentence = " ".join(sentence.split())  # Remove extra spaces
        sentence = sentence.lower()  # Convert to lowercase
        clean_sentences.append(sentence)

    # Handle empty sentences (optional)
    if not clean_sentences:
        return ""  # Return empty string if no clean sentences

    # Join cleaned sentences into a single string
    joined_text = " ".join(clean_sentences)

    return joined_text
```

```python
clean_corpus = []
for text in corpus:
  clean_corpus.append(clean_text(text))
```

```python
from spacy.lang.id import Indonesian
import numpy as np
spacy_id = Indonesian()

all_tokens = []
for sentence in clean_corpus:
    word_token = spacy_id(sentence)
    tokens = [token.text for token in word_token]
    all_tokens.append(tokens)
```

## Start Query Expansion
### using latent semantic analysis (LSA)

```python
df_id = pd.DataFrame({'document': clean_corpus})
df_id.head()
```

```python
from gensim import corpora, models, similarities

# Create a dictionary from the tokenized documents
dictionary = corpora.Dictionary(all_tokens)
```

```python
# Convert tokenized documents to a document-term matrix
corpus = [dictionary.doc2bow(doc) for doc in all_tokens]
```

```python
# Build an LSA model
lsa_model = models.LsiModel(corpus, id2word=dictionary, num_topics=2)
```

```python
# Transform the documents to the LSA space
lsi_corpus = lsa_model[corpus]
```

```python
judulartikel = df['judul']
```

```python
def main_query(query_in):
    # Transform the query to the LSA space
    query_bow = dictionary.doc2bow(query_in.lower().split())
    query_lsi = lsa_model[query_bow]

    # Calculate similarity between the query and documents
    index = similarities.MatrixSimilarity(lsi_corpus)
    sims = index[query_lsi]

    # Sort the similarity scores
    sims_sorted = sorted(enumerate(sims), key=lambda item: -item[1])

    # Print the top matching documents
    print("Top matching documents:")
    for doc_id, score in sims_sorted:
        print(f"Document ID: {doc_id}, Score: {score}, Document: {judulartikel[doc_id]}")
```
